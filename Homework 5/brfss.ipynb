```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('brfss.csv')
print(data)
```

```python
list1 = data['weight2']
print("data",list1)
```

```python
import matplotlib.pyplot as plt

list1 = data['weight2']
list2 = data['wtyrago']
list3 = data['htm3']

plt.figure(figsize = (10,6))
plt.scatter(x=[1,2,3],y=[np.max(list1),np.max(list2),np.max(list3)], marker='v', color='black', label='max')
plt.scatter(x=[1,2,3],y=[np.min(list1),np.min(list2),np.min(list3)], marker='^', color='black', label='min')
plt.scatter(x=[1,2,3],y=[np.mean(list1),np.mean(list2),np.mean(list3)], marker='+', color='blue', label='mean')
plt.scatter(x=[1,2,3],y=[np.std(list1),np.std(list2),np.std(list3)], marker='x', color='blue', label='std')
plt.scatter(x=[1,2,3],y=[np.mean(list1) + np.std(list1),np.mean(list2)+ np.std(list2) ,np.max(list3) + np.std(list3)], marker='v', color='green', label='mean + std')
plt.scatter(x=[1,2,3],y=[np.mean(list1) - np.std(list1),np.mean(list2) - np.std(list2) ,np.max(list3) - np.std(list3)], marker='v', color='green', label='mean - std')
plt.scatter(x=[1,2,3],y=[np.percentile(list1,75),np.percentile(list2,75),np.percentile(list3,75)], marker='v', color='red', label=' 75% percentile')
plt.scatter(x=[1,2,3],y=[np.percentile(list1,25),np.percentile(list2,25),np.percentile(list3,25)], marker='^', color='red', label=' 25% percentile')
plt.xlim([0,4])
plt.xticks([1,2,3],['current weight','weight one year ago','height'])
plt.legend(loc = 'best')
plt.title('Exercice 1: Summary of the current weight, weight  one year ago and height')
plt.rcParams['axes.unicode_minus'] = False
plt.show()
```

```python
weight2 = data['weight2']
wtyrago = data['wtyrago']
age = data['age']
weight_change = weight2 - wtyrago
print(weight_change)

result1 = np.corrcoef(weight2, weight_change)
print(result1)
result2 = np.corrcoef(wtyrago,weight_change)
print(result2)
result3 = np.corrcoef(age,weight_change)
print(result3)
```

```python
plt.scatter(x = data['weight2'], y = data['weight_change'])
plt.title('current weight - weight change')
plt.show
```

```python
plt.scatter(x = wtyrago, y = weight_change)
plt.title('weight one year - weight change')
plt.show
```

```python
plt.scatter(x = data['age'], y = data['weight_change'])
plt.title('age - weight change')
plt.show
```
